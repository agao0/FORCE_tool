```python
import pandas
import numpy as np
from matplotlib import pyplot as plt 
```

```python
graphs = []

f = open("build/tests/errors.txt", "r")
curr_index = -1

for x in f:
    if x.strip() == '':
        continue
    if x.startswith("Run"):
        continue
    if x.startswith("THETAD"):
        num = float(x.split(":")[1])
        curr = [[[num], [], [], [], [], [], [], [], [], [], []]]
        curr_index += 1
        graphs += curr

        continue
        
    num = float(x.split(":")[1])
    if x.startswith("LLBJ"):
        graphs[curr_index][1] += [num]
    if x.startswith("LUBJ"):
        graphs[curr_index][2] += [num]
    if x.startswith("LWBJ"):
        graphs[curr_index][3] += [num]
    if x.startswith("LInertia"):
        graphs[curr_index][4] += [num]
    if x.startswith("LCG"):
        graphs[curr_index][5] += [num]
    if x.startswith("RLBJ"):
        graphs[curr_index][6] += [num]
    if x.startswith("RUBJ"):
        graphs[curr_index][7] += [num]
    if x.startswith("RWBJ"):
        graphs[curr_index][8] += [num]
    if x.startswith("RInertia"):
        graphs[curr_index][9] += [num]
    if x.startswith("RCG"):
        graphs[curr_index][10] += [num]

for i in range(len(graphs)):
    plt.style.use('seaborn-deep')
    plt.figure(figsize=(10,3))
    plt.title("THETAD UPPER: " + str(graphs[i][0]) +" LEFT ERRORS")
    plt.hist(graphs[i][1], label = "LLBJ")
    plt.hist(graphs[i][2], label = "LUBJ")
    plt.hist(graphs[i][3], label = "LWBJ")
    plt.hist(graphs[i][4], label = "LInertia")
    plt.hist(graphs[i][5], label = "LCG", bins = np.arange(0, 0.005, 0.0001))
#     plt.hist(graphs[i][5], label = "LCG", bins = 30)
#     plt.hist([x, y], bins, label=['x', 'y'])
    plt.legend(loc='upper right')
    plt.show()
print("Max error:" + str(max(graphs[i][5])))
```

```python
for i in range(len(graphs)):
    plt.style.use('seaborn-deep')
    plt.figure(figsize=(10,3))
    plt.title("THETAD UPPER: " + str(graphs[i][0]) +" RIGHT ERRORS")
    plt.hist(graphs[i][6], label = "RLBJ")
    plt.hist(graphs[i][7], label = "RUBJ")
    plt.hist(graphs[i][8], label = "RWBJ")
    plt.hist(graphs[i][9], label = "RInertia")
    plt.hist(graphs[i][10], label = "RCG", bins = np.arange(0, 0.007, 0.0001))
    plt.legend(loc='upper right')
    plt.show()
print("Max error:" + str(max(graphs[i][10])))
```
